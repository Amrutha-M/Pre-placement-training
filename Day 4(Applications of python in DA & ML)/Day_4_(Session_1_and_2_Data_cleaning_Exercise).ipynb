#Case Study 1

```
import pandas as pd 
import numpy as np 
import io

from google.colab import files 
uploaded = files.upload()
df = pd.read_csv(io.BytesIO(uploaded['property_data1.csv']))
df
```

#Checking for null values in all columns

```
print(df['ST_NUM'])
print(df['ST_NUM'].isnull())
print(df['NUM_ROOMS'])
print(df['NUM_ROOMS'].isnull())
print(df['OWN_OCCUPIED'])
print(df['OWN_OCCUPIED'].isnull())
```

#converting all the dirty values of 'NUM_ROOMS' column to NaN 

```
c=0
for i in df['NUM_ROOMS']:
  try:
    int(i)
  except ValueError:
    df.loc[c,'NUM_ROOMS']=np.nan
  c+=1
print(df['NUM_ROOMS'])
print(df['NUM_ROOMS'].isnull())
```

#converting all the dirty values of 'OWN_OCCUPIED' column to NaN 

```
c=0
for i in df['OWN_OCCUPIED']:
  try:
    int(i)
    df.loc[c,'OWN_OCCUPIED']=np.nan
  except ValueError:
    pass
  c+=1
print(df['OWN_OCCUPIED'])
print(df['OWN_OCCUPIED'].isnull())
```

#Finding out sum of null values in each column and in whole dataframe

```
print(df.isnull().sum())
print(sum(df.isnull().sum()))
```

#Printing average values of all rows

```
print(df['ST_NUM'].isnull().sum()/9.0)
print(df['ST_NAME'].isnull().sum()/9.0)
print(df['NUM_ROOMS'].isnull().sum()/9.0)
print(df['OWN_OCCUPIED'].isnull().sum()/9.0)
```

#For values greater than 10% it tells it still contains dirty data or nan values which should be replced by valid values

```
df['ST_NUM'].fillna(120,inplace=True)
print(df['ST_NUM'])
m = df['NUM_ROOMS'].median()
df['NUM_ROOMS'].fillna(m,inplace=True)
print(df['NUM_ROOMS'])
```

#Checking highest frequency value in 'OWN_OCCUPIED'

```
df['OWN_OCCUPIED'].value_counts().idxmax()
```

#Replacing NaN values with 'N' in 'OWN_OCCUPIED'

```
df['OWN_OCCUPIED'].fillna('N',inplace=True)
print(df['OWN_OCCUPIED'])
```

```
df['ST_NAME'].value_counts().idxmax()
```

```
df['ST_NAME'].fillna('BLAZE',inplace=True)
print(df['ST_NAME'])
```
